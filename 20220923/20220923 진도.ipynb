### Time Series

```python
import pandas as pd
```

```python
dates=pd.date_range("2016-05-01","2016-05-07")
dates
```

```python
tmp1=pd.Series([80,92,82,85,97,84,78],index=dates)
tmp1
```

```python
tmp2=pd.Series(np.random.randint(60,100,size=7),index=dates)
tmp2
```

```python
#데이터 프레임 만들기
exam=pd.DataFrame({
    "Math":tmp1,
    "philosophy":tmp2
})
exam
```

```python
#2016년 5월5일 점수만 보기
exam.loc["2016-05-05"]
```

```python
#5/4,5/6,5/7의 철학 점수만 보기
exam['philosophy'][[3,5,6]]
```

```python
#1 수학점수와 철학점수의 차 구하기 
# 새로운 열에 추가
```

```python
a=tmp1-tmp2
```

```python
exam["difference"]=a
exam
```

```python
# 컬럼 이름 바꾸기
exam.rename(columns={'Math':'Mathmatics',"difference":"diff"},inplace=True)
exam.columns
```

```python
exam
```

```python
#exam에서 'diff' 칼럼을 지워주새요
exam.drop('diff',axis=1,inplace=True)
exam.head()
```

```python
#and , or 중 하나의 사용(&,|)
#5/3,5/4의 시험 점수를 데이터 프레임 형태로  가져오세요!
exam[(exam.index=="2016-05-03")|(exam.index=="2016-05-04")]
```

```python
#and , or 중 하나의 사용(&,|)
#수학 점수가 85점 이상
#5/1~5/5
exam[(exam.Mathmatics>85)&(exam.index<'2016-05-05')]
#
```

```python
exam.plot()
```

```python
#바 차틀로 표현
exam.plot(title="Score",figsize=(15,5),kind="bar")
plt.xlabel("Date")
plt.ylabel("Score")
```

### 우리의 목표: 영화 상영시간이 200분 이상인 영화만 필터링 해보자!

```python
booleans=[]
for length in movies.duration:
    if length>=200:
        booleans.append(True)
    else:
        booleans.append(False)
```

```python
booleans
```

```python
#데이터 프레임을 만들기전에 리스트를 Series로 변환
is_long=pd.Series(booleans)
is_long
```

```python
movies[movies.duration>=200]
```

```python
# 각나라별 음주현황데이터 불러오기
drinks = pd.read_csv('http://bit.ly/drinksbycountry')
drinks.head()
```

```python
drinks[drinks.continent.isin(['Asia','Europe','Africa'])].head()
```

```python
drinks[drinks.beer_servings>200].head()
```

```python
#전체 데이터의 맥주 평균 소비량
b=drinks.beer_servings.mean()
#전체 데이터의 spirit 평균 소비량
c=drinks.spirit_servings.mean()
#전체 데이터의 와인 소비량
d=drinks.wine_servings.mean()
```

```python
# 아프리카 대륙의 평균 소비량
drinks[drinks.continent=="Africa"].wine_servings.mean()
```

```python
# 각 대륙별 맥주 최고 소비량
drinks.groupby('continent').beer_servings.max()
```

```python
#다양한 통계치를 한꺼번에!
drinks.groupby('continent').beer_servings.agg(["count","mean","min","max"])
```

```python
drinks.groupby('continent').mean().plot(kind="bar")
```
