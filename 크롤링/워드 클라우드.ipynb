```python
def extract_tag(string):
    result_words=[]
    result = okt.pos(string, stem=True, norm=True)
    for word, tag in result:
        if tag in['Noun','Verb']:
            if word not in stopwords:
                result_words.append(word)
    return result_words
```

```python
from konlpy.tag import Okt
okt = Okt()
```

```python
stopwords=['있다.','하다','되다','고','말','며','등','것','이',"전"]
```
