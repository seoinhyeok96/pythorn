```python
price=soup.select("td.number")
```

```python
len(price)
```

```python
price_raw_list=[]
for e  in price:
    e.text
    print(e.text)
    price_raw_list.append(e.text)
```

```python
price_raw_list
```

```python
price_list=[]
for e in range(0,1000):
    if e % 10 ==0:
        price= price_raw_list[e]
        price_list.append(price)
```
