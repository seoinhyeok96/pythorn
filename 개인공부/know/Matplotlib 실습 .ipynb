### 다음과 같이 X,Y 값이 주어 졌을때, 이를 빨간색, 수평막대 그래프를 그려보세요!



```python
x = [1, 2, 3, 4]
y = [5, 4, 3, 2]

# X와 Y를 내림차순으로 정렬합니다.
x_descending = sorted(x, reverse=True)
y_descending = sorted(y, reverse=True)

# Figure와 Axes를 생성합니다.
fig, ax = plt.subplots()

# 수평 막대 그래프를 그립니다.
ax.barh(x_descending, y_descending, color='red')

# X축과 Y축 레이블을 설정합니다.
ax.set_xlabel('Y')
ax.set_ylabel('X')

# 그래프를 출력합니다.
plt.show()
```

## 'abcdfg'라는 문자열에서 'cde'라는 문자열을 뽑아 내고 싶다. 정규표현식으로 코드를 구현해 보라 이때 '.'를 활용해 구현해 보라

```python
import re

text = 'abcdfg'

pattern = r'cde'
match = re.search(pattern, text)

if match:
    result = match.group()
    print(result)
else:
    print('No match')
```

## 'abbcdeeee'라는 문자열이 있다. 여기서 'abbc'가 뽑히도록 정규표현식을 이용해 코딩해 보라 정규 표현식에 {}가 들어가도록 해보아라

```python
import re

text = 'abbcdeeee'

pattern = r'ab{2}c'
match = re.search(pattern, text)

if match:
    result = match.group()
    print(result)
else:
    print('No match')
```
