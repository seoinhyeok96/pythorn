```python
# 기울기가 2 , y절편 3인 회귀 곡선 그려보기
# x의 범위는 0~10
# random.gauss(평균,표준 편차)
x1=[]
y1=[]
for i in range(1000):
    imsi = random.random()*10
    x1.append(imsi)
    y1.append(2*imsi+3+random.gauss(0,2))
plt.scatter(x1,y1)
```

```python
x2= range(11)
y2=[2*n+3 for n in range(11)]
plt.plot(x2,y2)
```

```python
plt.scatter(x1,y1,c='black')
plt.plot(x2,y2, c= "blue")
plt.show()
```

```python
X=sm.add_constant(x1)
model= sm.OLS(y1,X)
result= model.fit()
print(result.summary())
```

```python
x4=[]
x4_1=[]
y4=[]
for i in range(1000):
    imsi=random.random()*10
    x4.append(imsi)
    x4_1.append(imsi+ random.gauss(0,1))
    y4.append(2*imsi + random.gauss(3,2))
df={"x4":x4,"x4_1":x4_1,"y4":y4}
df=pd.DataFrame(df)

x4= df[["x4","x4_1"]]
y4= df["y4"]

x4=sm.add_constant(x4, has_constant = "add")
model= sm.OLS(y4,x4)
result = model.fit()

print(result.summary())
```
